```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=300
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
```

```python
data = pd.read_csv("sp500NoNA_daily_simple_1000.csv", encoding = 'gbk')
data.head()
```

```python
data = data.drop(columns = ['Date'])
```

```python
data
```

```python
pca150 = PCA(n_components=150)
pca_result_150 = pca150.fit_transform(data)
# 前100个主成分的解释方差比之和
explained_variance_sum = pca150.explained_variance_ratio_.sum()
print(explained_variance_sum)
```

```python
df_pca150 = pd.DataFrame(pca_result_150, columns=[f'PC{i+1}' for i in range(pca_result_150.shape[1])])
df_pca150.to_csv('pca150.csv', index=False)
```
